# ggplot 2 

```r
#importing the library
library("tidyverse")
```

```r
# reading the dataset
df<-iris
```

```r
df
```

```r
#plotting
plot1<-ggplot(df,aes(Sepal.Length,Petal.Length,color=Species))+geom_point()
print(plot1)
```

# Changing the x axis and the y axis as well as giving the plot a title

```r
print(plot1+labs(y="petal length(cm)",x="Sepal length (cm)")+ggtitle("Petal and sepal length of iris"))
```

# highlighting an area in the plot

```r
plot1+annotate("rect",xmin=5,xmax=7,ymin=4,ymax=6,alpha=.3)
```

# segmenting the plot

```r
plot1+annotate("segment",x=5,xen=7,y=4,yend=5,color="red")
```

# removing the legend

```r
plot1+theme(legend.position="none")
```

# hiding the legend title

```r
plot1+theme(legend.title=element_blank())
```

# changing the legend position

```r
plot1+theme(legend.position="top")
```

# using geom_smooth (using method=lm gives the regression line along with the confidence region)

```r
plot1+geom_smooth(method=lm)
```

# jitter plot

```r
df2<-mpg
```

```r
ggplot(df2,aes(cyl,hwy))+geom_point()+geom_jitter(aes(color=class))
```

# bar plots and histograms

```r
# bar plot
ggplot(df2,aes(x=factor(cyl)))+geom_bar(stat="count")
```

```r
#histogram
ggplot(df2,aes(x=hwy))+geom_histogram(col="red",fill="green",alpha=.6,binwidth=3)
```

```r
# stacked bar chart
plot2<-ggplot(df2,aes(class))
plot2+geom_bar()
```

# pie charts

```r
df<- as.data.frame(table(mpg$class))
```

```r
colnames(df)<-c("class","freq")
df
```

```r
pie<-ggplot(df,aes(x="",y=freq,fill=factor(class)))+geom_bar(width=1,stat="identity")+theme(axis.line=element_blank(),plot.title=element_text(hjust=0.5))+labs(fill="class",x=NULL,y=NULL,title="pie chart of class",caption="source: mpg")
```

```r
print(pie+coord_polar(theta="y",start=0))
```

```r
p<-ggplot(mpg,aes(cty))+geom_density(aes(fill=factor(cyl)), alpha=0.8)
```

```r
p+labs(title="density plot",subtitle="City Milage Grouped by Number of Cylinders",caption="Source: mpg",x="City Milage",fill="# Cylinders")
```

# Box plot

```r
p<-ggplot(mpg, aes(class, cty))+geom_boxplot(varwidth=T)
p+labs(title="A Box plot Example",subtitle="Mileage by Class",caption="MPG Dataset",x="Class",y="Mileage")
```

# Dot plot

```r
p<-ggplot(mpg,aes(manufacturer,cty))
p+geom_dotplot(binaxis='y',stackdir='center',dotsize=.7)
```

# violin plot

```r
p<-ggplot(mpg,aes(class,cty))+geom_violin()
```
