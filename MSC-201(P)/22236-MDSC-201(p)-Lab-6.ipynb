# Importing the CSV file

```r
data<-read.csv('rm.csv')
print(data)
```

# Analysing the CSV file

```r
print("Checking if the CSV file is imported as a dataframe?")
print(is.data.frame(data))
print("No of columns in the dataframe: ")
print(ncol(data))
print("No of rows in the dataframe: ")
print(nrow(data))
```

# getting the maximum salary

```r
sal<-max(data$salary)
print(sal)
```

# getting the details of the person with the maximum salary

```r
person<-subset(data,salary==sal)
print(person)
```

# getting all the people working in the IT dept

```r
people<-subset(data,dept=="IT")
people
```

# getting the people in IT dept whose salary is greater than 600

```r
people<-subset(data,salary>600 & dept=="IT")
print(people)
```

# people who joined on or after 2014

```r
people<-subset(data,as.Date(start_date)>as.Date("2014-01-01"))
print(people)
```

# importing the data

```r
passengers<-read.csv("./train.csv")
```

```r
passengers
```

# checking the summary of the data

```r
summary(passengers)
```

# using the pipe operator

```r
passengers %>%
    summary()
```

# dropping the null values

```r
passengers %>%
    drop_na() %>%
    summary()
```

# wrangling the data
* includes the operations such as
    * filter()
    * arrange()
    * mutuate()
    * these are from the dplyr package which is included in the tidyverse package

```r
# filtering to get all the male rows
passengers %>%
    filter(Sex=="male")
```

```r
# filter to get all the female rows
passengers %>%
    filter(Sex=="female")
```

```r
# arrange by increasing fare
passengers %>%
    arrange(Fare)
```

```r
# arrange by decreasing fare
passengers %>%
    arrange(desc(Fare))
```

```r
# creating a new column FamSize (size of family)
passengers %>%
    mutate(FamSize = Parch+SibSp)
```

```r
# arranging the new column in a decreasing order
passengers %>%
    mutate(FamSize=Parch+SibSp) %>%
    arrange(desc(FamSize))
```

```r
# turning the numerical values of Survived column to "No" & "Yes" (new dataframe)
passengers1<-passengers %>%
    mutate(Survived=ifelse(Survived==0,"No","Yes"))
```

```r
passengers1
```

# plotting the data thus we created

```r
ggplot(passengers,aes(x=Sex))+geom_bar()
```

```r
# scatter plot of age vs fare
ggplot(passengers,aes(x=Age,y=Fare))+geom_point()
```

```r
# scatter plot of age vs fare colored by sex
ggplot(passengers,aes(x=Age,y=Fare,color=Sex))+geom_point()
```

```r
# scatter plot of age vs fare colored by sex facated by survived
ggplot(passengers1,aes(x=Age,y=Fare,color=Sex))+geom_point()+facet_grid(~Survived)
```

```r
# plot barplot of passengers Sex and fill according to Survival 
ggplot(passengers1,aes(x=Sex,fill=Survived))+geom_bar()
```

# Summarizing and grouping the data
* using the methods 
    * summarize()
        * mean
        * median

```r
# checking the mean fare
passengers %>%
    summarize(meanFare=mean(Fare))
```

```r
# checking the median of the Fare
passengers %>%
    summarise(medianFare=median(Fare))
```

```r
# using the filter method to check the average fare paid by male passsengers
passengers %>%
    filter(Sex=="male") %>%
    summarise(meanFare=mean(Fare))
```

```r
# using the filter method to check the average fare paid by female passengers
passengers %>%
    filter(Sex=="female") %>%
    summarize(meanfare=mean(Fare))
```

```r
# Check out mean Fare & number of survivors for women
passengers %>%
  filter(Sex == "female") %>%
  summarise(meanFare = mean(Fare), numSurv = sum(Survived))
```

```r
# Check out mean Fare & number of survivors grouped by Sex
passengers %>%
  group_by(Sex) %>%
  summarise(meanFare = mean(Fare), numSurv = sum(Survived))
```

```r
# Check out mean Fare & proportion of survivors grouped by Sex
passengers %>%
  group_by(Sex) %>%
  summarise(meanFare = mean(Fare), numSurv = sum(Survived)/n())
```
