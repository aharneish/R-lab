```r
#importing the data set
data<-mtcars
head(data)
```

# Pie Charts

```r
#pie chart
x<-data$cyl
y<-data$carb
pie(x,y,radius=1,main="cylinders vs carborator",clockwise=TRUE)
```

# Box Plots

```r
#box plots
boxplot(hp~gear,data=data,xlab="gear",ylab="horsepower",main="horsepower vs gear")
```

# Bar Plot

```r
# BAR PLOT
s<-table(data$cyl)
print(s)
barplot(s,main="No of cylinders",xlab="cylinders",ylab="no of cylinders",border="black",col="red")
```

# Mean

```r
val<-c(100:180)
m<-mean(val)
print(m)
```

```r
# mean calculated manually
mean_cal=0
for(i in val)
{
  mean_cal=mean_cal+i
}
mean_cal=mean_cal/length(val)
print(mean_cal)
```

# Mode

```r
mode<-function(val)
{
    u<-unique(val)
    u[which.max(tabulate(match(val,u)))]
}
```

```r
val<-c(10,0,0,2,30,30,30,30,19,1,2,3,4,50)
print(mode(val))
```
